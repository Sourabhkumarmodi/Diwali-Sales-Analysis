# Exploratory Data Analysis

```python
df['Gender'].unique()
```

```python
df['Gender'].value_counts()
```

```python
# ploting the barchart for Gender and it's count
ax=sns.countplot(df['Gender'])
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
#Conclusion: female can purchase more the male
```

```python
sns.boxplot(y='Age',x='Age Group',hue='Gender',data=df)
```

```python
ax=sns.countplot(x='Age Group',hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```
